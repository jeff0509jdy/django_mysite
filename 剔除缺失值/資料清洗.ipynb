```python
#https://zhuanlan.zhihu.com/p/32572237
#https://zhuanlan.zhihu.com/p/93179647
import pandas as pd
import numpy as np
import datetime
```

```python
df01 = pd.read_excel("價格資料_論文用_1e(未剔除缺失值之資料)_0831.xlsx")
df01
```

```python
#判斷行或列是否有缺失值 https://blog.csdn.net/crj0926/article/details/100805102
# https://blog.csdn.net/qq_40825479/article/details/83544430

#https://zhuanlan.zhihu.com/p/89259675
def check_missing_data(df):
    # check for any missing data in the df (display in descending order)
    return df.isnull().sum(axis = 1).sort_values(ascending=False)
check_missing_data(df01)
```

```python
#讀取任一列的數字 iloc輸入值為索引值；loc是基於標籤
#https://www.cnblogs.com/timotong/p/9678490.html
df01.iloc[1096]
```

```python
# 去除有NaN值的行或列(axis=0去除列，=1去除行
df01.dropna(axis = 0, inplace = True)
df01
```

```python
# https://www.delftstack.com/zh-tw/howto/python-pandas/pandas-remove-index/
# 重置 DataFrame 的索引
df01 = df01.reset_index(drop=True)
df01
```

```python
# 將 DataFrame 的索引改設置為 Date
df01 = df01.set_index('Date')
df01
```

```python
#https://blog.csdn.net/Benzmjs/article/details/116598639
df01.to_excel("價格資料_論文用_2e(剔除缺失值之資料)_0831.xlsx", index = False)
```
